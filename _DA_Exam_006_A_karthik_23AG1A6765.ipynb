```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import re
data = pd.read_csv("/content/car_price_performance.csv")
data.head()
```

```python
brand_avg_prices = data.groupby('Brand')['Price'].mean().sort_values()
brand_mapping = {brand: i for i, brand in enumerate(brand_avg_prices.index)}
data['Brand_encoded'] = data['Brand'].map(brand_mapping)
print(data[['Brand', 'Brand_encoded']].head())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
data[['Horsepower', 'Fuel Efficiency']] = scaler.fit_transform(data[['Horsepower', 'Fuel Efficiency']])
data['Brand_encoded'] = scaler.fit_transform(data[['Brand_encoded']])
print(data[['Horsepower', 'Fuel Efficiency', 'Brand_encoded']].head())
```

```python

data['Performance Index'] = (data['Horsepower'] * 0.7 + data['Fuel Efficiency'] * 0.3) / data['Price']

print(data.head())
```

```python

result_df = data[['Brand', 'Price', 'Horsepower', 'Fuel Efficiency', 'Brand_encoded']]
result_df = result_df.rename(columns={'Horsepower': 'Scaled Horsepower', 'Fuel Efficiency': 'Scaled Fuel Efficiency', 'Brand_encoded': 'Scaled Brand'})
result_df['Performance Index'] = data['Performance Index']
result_df
```
